# Bootstrap Bodyfat

## Bootstrap with matplotlib

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('dark_background')

data = pd.read_csv('datasets/bodyfat.csv')

bootstrap = []
N = 50
for n in range(N):  
    sample = data.sample(30, replace=True)
    bootstrap.append(sample.mean())
df_bootstrap = pd.DataFrame(bootstrap)

print('Bootstrap Statistics:')
print(f'original: {data.mean().to_string()}')
print(f'bootstrap: {df_bootstrap.mean().to_string()}')
print(f'bias: {df_bootstrap["Weight"].mean()- data["Weight"].mean()}')
print(f'std.error: {df_bootstrap.std().to_string()}')

fig = plt.figure(figsize =(12, 8))
ax = fig.add_subplot(111)
 
bp = ax.boxplot(df_bootstrap[["Weight", "Abdomen", "Thigh"]], 
                patch_artist = True,
                vert = 0)

ax.set_yticklabels(["Weight", "Abdomen", "Thigh"])
 
plt.title("BodyFat")

plt.show()
```

## Bootstrap with pandas

```python
import pandas as pd

data = pd.read_csv('datasets/bodyfat.csv')

bootstrap = []
N = 120
for n in range(N):  
    sample = data.sample(20, replace=True)
    bootstrap.append(sample.mean())
df_bootstrap = pd.DataFrame(bootstrap)

print('Bootstrap Statistics:')
print(f'original: {data.mean().to_string()}')
print(f'bootstrap: {df_bootstrap.mean().to_string()}')
print(f'bias: {df_bootstrap["Weight"].mean()- data["Weight"].mean()}')
print(f'std.error: {df_bootstrap.std().to_string()}')

bootstrap2 = []
N = 20
for n in range(N):  
    sample = data.sample(5, replace=True)
    bootstrap2.append(sample.mean())
df_bootstrap2 = pd.DataFrame(bootstrap2)

df_bootstrap[["Weight", "Abdomen", "Thigh", "Chest"]].plot.box(
    vert=False, 
    showmeans=True, 
    showfliers=True, 
    positions=[1, 2, 4, 3])

print('Bootstrap Statistics 2:')
print(f'original: {data.mean().to_string()}')
print(f'bootstrap: {df_bootstrap2.mean().to_string()}')
print(f'bias: {df_bootstrap2["Weight"].mean()- data["Weight"].mean()}')
print(f'std.error: {df_bootstrap2.std().to_string()}')

df_bootstrap2[["Weight", "Abdomen", "Thigh", "Chest"]].plot.box(
    vert=False, 
    showmeans=True, 
    showfliers=True, 
    positions=[1, 2, 4, 3])
```
