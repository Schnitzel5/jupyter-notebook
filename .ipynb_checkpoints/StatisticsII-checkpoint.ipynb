# Data and Sampling Distributions

### Sample
- A subset from a larger data set.
###  Population
- The larger data set or idea of a data set.
###  $N(n)$ 
- The size of the population (sample).
###  Random sampling
- Drawing elements into a sample at random.
###  Simple random sample
- The sample that results from random sampling without stratifying the population.
### Bias
- Systematic error.
### Sample bias
- A sample that misrepresents the population.

```python
import pandas as pd
import seaborn as sns
import sklearn
from sklearn.utils import resample
import matplotlib as plt
import numpy as np

loans_income = pd.read_csv("c:/home/DSAI/datasets/loans_income.csv")
results = []
for nrepeat in range(1000):
    sample = resample(loans_income)
    results.append(sample.median())
    
results = pd.Series(results)
print('Bootstrap Statistics:')
print(f'original: {loans_income.median()}')
print(f'bias: {results.mean() - loans_income.median()}')
print(f'std. error: {results.std()}')
```

## Bootstraping
#### I statistics, Bootstrap Sampling is a method that involves drawing of sample data repeatedly with replacement from a data source to estimate a population parameter
#### sample() Method for bootstraping 

```python
sample_data = loans_income.sample(1000,replace=True)
sample_data_m = loans_income.sample(1000).mean()
sample_mean_05 = [loans_income.sample(5).mean() for _ in range(1000)]
sample_mean_05 = pd.DataFrame(sample_mean_05)
sample_mean_20 = [loans_income.sample(20).mean() for _ in range(1000)]
sample_mean_20 = pd.DataFrame(sample_mean_20)
sample_mean_100 = [loans_income.sample(100).mean() for _ in range(1000)]
sample_mean_100 = pd.DataFrame(sample_mean_100)
                                  
                                   
```

###  Sample statistic
A metric calculated for a sample of data drawn from a larger population.
###  Data distribution
The frequency distribution of individual values in a data set.
###  Sampling distribution
The frequency distribution of a sample statistic over many samples or resamples.
###  Central limit theorem
The tendency of the sampling distribution to take on a normal shape as sample
size rises.
###  Standard error
The variability (standard deviation) of a sample statistic over many samples (not
to be confused with standard deviation, which by itself, refers to variability of
individual data values).

```python
fig, axs = plt.pyplot.subplots(1, 4, figsize=(12, 4)) #constrained_layout=True)
#plt.subplots_adjust(hspace=0.5)
b = 15
sample_data.hist(bins=b, ax=axs[0],color="r")
sample_mean_05.hist(bins=b, ax=axs[1],color="g")
sample_mean_20.hist(bins=b, ax=axs[2],color="y")
sample_mean_100.hist(bins=b, ax=axs[3])
```

## Confidence Intervals
### Confidence level
The percentage of confidence intervals, constructed in the same way from the
same population, that are expected to contain the statistic of interest.
## Interval endpoints
The top and bottom of the confidence interval.

```python
ul = np.percentile(sample_data,2.5)
ol = np.percentile(sample_data,97.5) 
print(ul,ol)
sample_data.hist(bins=20)
plt.pyplot.axvline(ul, color='r', linestyle='dashed', linewidth=1)
plt.pyplot.axvline(ol, color='r', linestyle='dashed', linewidth=1)
```

```python
x = np.random.normal(0, 1, 1000)
x = pd.DataFrame(x)
ul = np.percentile(x,2.5)
ol = np.percentile(x,97.5) 
print(ul,ol)
x.hist(bins=20)
plt.pyplot.axvline(ul, color='r', linestyle='dashed', linewidth=1)
plt.pyplot.axvline(ol, color='r', linestyle='dashed', linewidth=1)
```
